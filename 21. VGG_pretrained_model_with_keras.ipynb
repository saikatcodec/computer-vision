```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
from tensorflow.keras.utils import load_img, img_to_array
import numpy as np
```

### Load the VGG16 model

```python
model = VGG16(weights='imagenet')
model.summary()
```

### Load the VGG19

```python
from tensorflow.keras.applications import VGG19
```

```python
model = VGG19(weights='imagenet')
model.summary()
```

### Inferencing the model

```python
def predict(image_path):
  img = load_img(image_path, target_size=(224, 224))
  input = img_to_array(img)
  input = np.expand_dims(input, 0)

  processed = preprocess_input(input)
  predict = model(processed)

  result = decode_predictions(predict)

  return result[0]
```

```python
img_path = '/content/FS1325-1-big.jpg'
res = predict(image_path=img_path)

for i, (_id, class_name, score) in enumerate(res):
  print(f'{i + 1}. {class_name}: {score:.2f}')
```
