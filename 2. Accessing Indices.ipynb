````python
import torch
````

## Accessing tensor
```
tensor_name[row, col]
```

````python
# Create a tensor
tensor = torch.tensor([[1, 2, 3],
                       [4, 5, 6],
                       [56, 32, 21]])
````

### 1. Indexing

````python
# access 0th row and 1st column element
tensor[0, 1]
````

````python
# access 1st row
tensor[1]
````

````python
# access the 1st column
tensor[:, 1]
````

### 2. Slicing

````python
print(tensor)
# slice 0th, 1st row and 1st, 2nd column
tensor[:2, 1:3]
````

````python
# access element in step 2
# access 0th, 2nd row and all column
tensor[::2]
````

````python
# access all row and 0th, 2nd column
tensor[:, ::2]
````

````python
# access 1st row and 1st, 2nd column only
tensor[1, 1:3]
````

### Fancy Indexing

````python
tensor
````

````python
# access 0th row, 0th column and 2th row, 2th column
# (0, 0) ---> 1
# (2, 2) ---> 21
tensor[[0, 2], [0, 2]]
````

### Boolean Indexing

````python
mask = tensor > 30
print(mask)
print(tensor[mask])
````

### Changing Tensor values

````python
print('Before', tensor)

# Change 0th row, 1th column
tensor[0, 1] = 33

print('After', tensor)
````

````python
# Change 1st row value
tensor[1] = torch.tensor([12, 45, 76])
print(tensor)
````

````python
# Change 1st column
tensor[:, 1] = torch.tensor([99, 55, 21])
print(tensor)
````

### Select indices using function
0 for row<br>
1 for column

````python
indices = torch.tensor([0, 2])
selected_row = torch.index_select(tensor, 0, indices)
selected_row
````

````python
selected_column = torch.index_select(tensor, 1, indices)
selected_column
````
