````python
import torch
````

## Tensor Datatype
```
floating -- used for most the deep learning case ---> float32(default), float64, float16
integer -- used for some classification --> int32, int64(default), int8
boolean -- used for masking or indexing ---> bool
complex -- for complex and heavy computation ---> complex64, complex128
```

### Float number

````python
float_number = torch.tensor([1., 2.3, 4.5, 3.3])
float_number.dtype
````

````python
float_number = torch.tensor([1., 2.3, 4.5, 3.3], dtype=torch.float64)
float_number.dtype
````

### Interger number

````python
int_number = torch.tensor([1, 2, 3, 4])
int_number.dtype
````

````python
int_number = torch.tensor([1, 2, 3, 4], dtype=torch.int32)
int_number.dtype
````

### Boolean

````python
bool_data = torch.tensor([True, False, True, False, False])
bool_data.dtype
````

## Conversion

````python
# Float to integer
int_num = torch.tensor([1.2, 12.34, 5.5], dtype=torch.int64)
print(int_num)
int_num.dtype
````

### Another way

````python
int_number = float_number.to(torch.int64)
int_number
````

````python
# integer to boolean
bool_data = torch.tensor([0, -2, 3, 4, 0 , 0], dtype=torch.bool)
print(bool_data)
bool_data.dtype
````

````python
# integer to float
float_num = int_num.to(torch.float32)
print(float_num)
````
