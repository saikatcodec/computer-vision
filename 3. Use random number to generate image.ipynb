```python
import torch
import matplotlib.pyplot as plt
```

#### Random number betweeen 0 to 1

```python
# rand() for floating point number
torch.rand(4)
```

```python
torch.rand((4, 5))
```

```python
# Apply seed for random number
torch.manual_seed(32)

torch.rand(2, 3)
```

### Random Interger between low and high
[low, high)

```python
torch.randint(1, 10, (4, )) # generate number between 1 and 9
```

## Create noise image

```python
# Give hxw for grayscale image
noise = torch.rand((256, 256))

plt.imshow(noise, cmap='gray')
plt.title('Noise Image')
plt.axis('off')
plt.show()
```

## Create RGB noise image

```python
# Give HxWxC for rgb
rgb_noise = torch.rand((256, 256, 3))

plt.imshow(rgb_noise)
plt.title('Noise RGB Image')
plt.axis('off')
plt.show()
```
