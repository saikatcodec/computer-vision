```python
import torch
```

```python
torch.cuda.is_available()
```

### Creating Tensor

```python
# 0D tensor is called scaler
scaler = torch.tensor(7)

print('Scaler:', scaler)
print('Scaler size:', scaler.shape)
print('Scaler dimention:', scaler.ndim)
```

```python
# 1D tensor is called vector
vector = torch.tensor([1, 2, 3, 4])

print('vector:', vector)
print('vector size:', vector.shape)
print('vector dimention:', vector.ndim)
```

```python
# 2D tensor is called matrix
matrix = torch.tensor([[1, 2, 3], [2, 3, 4]])

print('matrix:\n', matrix)
print('matrix size:', matrix.shape)
print('matrix dimention:', matrix.ndim)
```

```python
# 3D tensor
tensor = torch.tensor([[[1, 2, 3], [10, 45, 56]], [[12, 34, 55], [23, 56, 67]]])

print('tensor:\n', tensor)
print('tensor size:', tensor.shape)
print('tensor dimention:', tensor.ndim)
```

#### GPU Acceleration

```python
if torch.cuda.is_available():
  gpu_matrix = matrix.to('cuda')
  print(gpu_matrix)
else:
  print('GPU Acceleration is not available')
```

```python
# tensor properties

print('size:', matrix.size())
print('dimension:', matrix.ndim)

print('GPU:', matrix.device)
print('GPU:', gpu_matrix.device)

print('Is matrix run on GPU:', matrix.is_cuda)
print('Is matrix run on GPU:', gpu_matrix.is_cuda)

print('Number of element in tensor:', matrix.numel())
```
