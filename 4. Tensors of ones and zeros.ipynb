## Create Zero's

```python
torch.zeros((4))
```

```python
zeros = torch.zeros((4, 5))
zeros
```

```python
zeros + 5
```

## Create zeros from other variable shape

```python
rand_matrix = torch.randint(2, 33, (5, 6))
rand_matrix
```

```python
# create zeros_like from rand_matrix shapes
like_others = torch.zeros_like(rand_matrix)
like_others
```
